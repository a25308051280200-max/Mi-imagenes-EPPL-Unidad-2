* Objetivo:
Desarrollar un programa interactivo en Python que permita al usuario realizar cálculos geométricos básicos relacionados con el círculo y la esfera, mediante un menú de opciones fácil de usar.
* Descripción:
El programa deberá mostrar inicialmente las fórmulas matemáticas que se utilizarán para los cálculos. Posteriormente, presentará un menú con las siguientes opciones:
* Calcular el área de un círculo
* Calcular el perímetro (circunferencia) de un círculo
* Calcular el volumen de una esfera
* Salir del programa
* Importación: Se usa import math para obtener el valor preciso de π (math.pi).
Funciones: Cada cálculo (areacirculo, perimetrocirculo, volumenesfera) es una función independiente que pide el radio dentro de sí misma, tal como se solicitó.
* Menú: La función menu() solo se encarga de mostrar las opciones y devolver el número que el usuario escribió.
* Bucle while True: En la función main, este bucle mantiene el programa corriendo infinitamente hasta que el usuario elige la opción 4, momento en el que se ejecuta break para terminar.
*Condicionales if/elif/else: Se usan para decidir qué función llamar según la opción del menú.
Estilo: Se han añadido emojis y separadores visuales (*, =) para hacerlo más elegante y fácil de leer en la consola.

```python
# ==================================
# Problema de calculo geometrico
# Elaborado por : Alejandra Rivera Valenzuela
# Fecha 2026
# =====================================
import math

# ====================================
# FUNCION = Calcular El Area Del Circulo
# FORMULA = A = π * r²
# ====================================
def mostrar_formulas():
    """Muestra las fórmulas matemáticas utilizadas."""
    print("\n" + "="*40)
    print("📐 FÓRMULAS A UTILIZAR 📐")
    print("="*40)
    print("1. Área del Círculo:")
    print("   Fórmula: A = π * r²")
    print("   Donde 'r' es el radio.")
    print("\n2. Perímetro (Circunferencia) del Círculo:")
    print("   Fórmula: P = 2 * π * r")
    print("   Donde 'r' es el radio.")
    print("\n3. Volumen de la Esfera:")
    print("   Fórmula: V = (4/3) * π * r³")
    print("   Donde 'r' es el radio.")
    print("="*40 + "\n")

def areacirculo():
    """Calcula el área de un círculo solicitando el radio."""
    print("\n🔵 CÁLCULO DEL ÁREA DEL CÍRCULO")
    radio = float(input("Por favor, ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo con radio {radio} es: {area:.2f}")

def perimetrocirculo():
    """Calcula el perímetro de un círculo solicitando el radio."""
    print("\n🟢 CÁLCULO DEL PERÍMETRO DEL CÍRCULO")
    radio = float(input("Por favor, ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f}")

def volumenesfera():
    """Calcula el volumen de una esfera solicitando el radio."""
    print("\n🟠 CÁLCULO DEL VOLUMEN DE LA ESFERA")
    radio = float(input("Por favor, ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f}")

def menu():
    """Muestra el menú principal y retorna la opción elegida."""
    print("\n" + "*"*30)
    print("🌟 MENÚ PRINCIPAL 🌟")
    print("*"*30)
    print("1️⃣  Calcular Área del Círculo")
    print("2️⃣  Calcular Perímetro del Círculo")
    print("3️⃣  Calcular Volumen de la Esfera")
    print("4️⃣  Salir del programa")
    print("*"*30)
    opcion = int(input("👉 Elige una opción (1-4): "))
    return opcion

# --- Programa Principal ---

def main():
    # Mostrar las fórmulas al inicio para referencia del estudiante
    mostrar_formulas()

    # Bucle principal para mantener el programa activo hasta que se elija salir
    while True:
        opcion_seleccionada = menu()

        if opcion_seleccionada == 1:
            areacirculo()
        elif opcion_seleccionada == 2:
            perimetrocirculo()
        elif opcion_seleccionada == 3:
            volumenesfera()
        elif opcion_seleccionada == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta luego. 🚀\n")
            break  # Sale del bucle while
        else:
            # Aunque no se pide validación compleja, este else maneja opciones fuera del rango 1-4
            print("\n⚠️  Opción no válida. Por favor, elige entre 1 y 4.")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
